```python
data["Type_of_order"]=label_encoder.fit_transform(data["Type_of_order"])
```

```python
data["City"]=label_encoder.fit_transform(data["City"])
```

```python
dummy1=pd.get_dummies(data.Festival)
dummy1
```

```python
dummy2=pd.get_dummies(data.Week_Time)
dummy2
```

```python
merged=pd.concat([data,dummy1,dummy2],axis='columns')
merged
```

```python
df=merged.drop(['No','Weekend'],axis='columns')
df
```

```python
df["Is_Festival"]=df["Yes"]
```

```python
df["Is_Weekday"]=df["Weekday"]
```

```python
fdf=df.drop(['Festival','Week_Time','Yes','Weekday'],axis='columns')
```

```python
fdf=fdf.drop(['Processing_Time','ID'],axis='columns')
```

```python
X=fdf.drop(['Time_taken_min'],axis='columns')
```

```python
Y=fdf['Time_taken_min']
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor=RandomForestRegressor(min_samples_leaf=48,n_estimators=1000, random_state=1)
```

```python
regressor.fit(X_Train,Y_Train)
```

```python
y_predtrain=regressor.predict(X_Train)
```

```python
from sklearn import metrics
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_Train, y_predtrain)
```

```python
y_pred=regressor.predict(X_Test)
```

```python
r2_score(Y_Test,y_pred)
```
